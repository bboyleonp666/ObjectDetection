## Install CUDA v10.0 for using GPU
<pre>
<a ref = "https://developer.nvidia.com/cuda-toolkit-archive">&#9658; CUDA Legacy Releases</a>
<a ref = "https://developer.nvidia.com/cudnn">&#9658; cuDNN Home Page</a>
<font size = 4><i>
Please make sure the <b>CUDA</b> you install is <b>v10.0</b> since the Tensorflow only works with the following CUDA

The version of <b>Tensorflow</b> we are going to install is <b>2.0.0</b>

Go to <b>environment variable</b> check if there are path created by installation, if not then add it manually

    <span style='background:Aquamarine'>Right click <b>Computer(本機)</b></span>
    -> <span style='background:Aquamarine'><b>Properties(內容)</b></span>
    -> <span style='background:Aquamarine'><b>Advanced system settings(進階系統設定)</b></span>
    -> <span style='background:Aquamarine'><b>Advanced(進階)</b></span>
    -> <span style='background:Aquamarine'><b>Environment variables(環境變數)</b></span>

    &bull; 1. variable name: CUDA_BIN
         variable value: C:\Program Files\NVIDIA GPU Computing Toolkit\CUDA\v10.0
         
    &bull; 2. variable name: CUDA_BIN_PATH
         variable value: C:\Program Files\NVIDIA GPU Computing Toolkit\CUDA\v10.0\bin

After install CUDA, for using gpu, we have to download the cuDNN that goes with CUDA, so pick the one

for CUDA v10.0, and then you should extract the files in cudnn.zip and put them into the directories of 

the same in the zip file
</i>

<b>Note:</b> In the cuDNN page, you have to sign up for download it</font></pre>

## Go to Anaconda Prompt
<pre>
<font size = 4><i>Type the command below will install Tensorflow GPU v2.0.0 automatically</i></font>

<font size = 4><i>If you just want the CPU ver. and skipped the part above you can ignore "-gpu" behind tensorflow</i></font>

    <span style='background:#E8E8E8'>conda -n <b>&#60;name&#62;</b> python=3.6 anaconda -y</span>
    <span style='background:#E8E8E8'>conda activate <b>&#60;name&#62;</b></span>

    <span style='background:#E8E8E8'>pip install astroid==2.3.3</span>
    <span style='background:#E8E8E8'>pip install spyder==3.3.6</span>
    <span style='background:#E8E8E8'>pip install --ignore-installed --upgrade tensorflow-gpu==2.0.0</span>
<font size = 4>
<b>Note:</b> - The <b>&#60;name&#62;</b> in the command line means the visual environment name which you can decide yourself
      - The python version installed here is 3.6.9 which is stable for tensorflow 2.0.0
      - conda module is only built under <b>&#60;base environment&#62;</b> so keep in mind that you cannot open 
        the prompt of <b>&#60;name&#62;</b> and try to use conda</font></pre>

## Install Tensorflow Object Detection API
<pre>
<font size = 4><i>You may download this repository either through "git clone" or just "download" it</i></font>

    <a ref = "https://github.com/tensorflow/models">&#9658; Tensorflow models repository</a>
    
    <font size = 4><b>Note:</b> One suggestion path of putting this repository is under <b>C:\Users\user\Tensorflow</b></font>

<font size = 4><i>Go to prompt and install necessary packages by pip if you have not installed them</i></font>

    <span style='background:#E8E8E8'>pip install cython</span>
    <span style='background:#E8E8E8'>pip install contextlib2</span>
    <span style='background:#E8E8E8'>pip install pillow</span>
    <span style='background:#E8E8E8'>pip install lxml</span>
    <span style='background:#E8E8E8'>pip install matplotlib</span>
    <span style='background:#E8E8E8'>pip install opencv-python</span>

<font size = 4><i>Install Protobufs for object detection api settings</i></font>

    <span style='background:#E8E8E8'>conda install protobuf -y</span>

<font size = 4><i>set environment variable again</i></font>

    &bull; 1. variable name: PYTHONPATH_MODEL
         variable value: <b>&#60;Path You Clone Models&#62;</b>\model
         
    &bull; 2. variable name: PYTHONPATH_RESEARCH
         variable value: <b>&#60;Path You Clone Models&#62;</b>\model\research
         
    &bull; 3. variable name: PYTHONPATH_SLIM
         variable value: <b>&#60;Path You Clone Models&#62;</b>\model\research\slim

<font size = 4><i>Head to <b>&#60;Path You Clone Models&#62;</b>\models\research and type the command below</i></font>

    <span style='background:#E8E8E8'>for /f %i in ('dir /b object_detection\protos\*.proto') do protoc object_detection\protos\%i --python_out=.</span>
    
<font size = 4><b>Note:</b> In <span style='background:Thistle'><b>&#60;Path You Clone Models&#62;</b>\models\research\object_detection\protos</span>,
      you will see lots of .py files appear after run the lines above correctly
</font>

<font size = 4><i>Under the same folder type</i></font>

    <span style='background:#E8E8E8'>python setup.py build</span>
    <span style='background:#E8E8E8'>python setup.py install</span>
    
<font size = 4><b>Note:</b> Your have to run these two lines every time you change anything of the object detection 
      package</font></pre>

<pre>
<font size = 7>
        <span style='background:red'>Everything is Done!!!</span>
</font>
</pre>

## Install labelImg
<pre>
<font size = 4><i>Open Prompt you and get into visual environment <b>&#60;name&#62;</b> you created</i></font>

<font size = 4><i>Install PyQt5 if you have not installed it by</i></font>

    <span style='background:#E8E8E8'>pip install pyqt5</span>

<font size = 4><i>If you follow this guide you will have the path <b>C:\Users\user\Tensorflow</b></i></font>

<font size = 4><i>Head to <b>C:\Users\user\Tensorflow</b> and create a folder named "addons" and move into it</i></font>

<font size = 4><i>Get labelImg by git clone or download as zip of the link below</i></font>

<a ref = "https://github.com/tzutalin/labelImg">&#9658; labelImg</a>

<font size = 4><i>Type command below to set it up</i></font>

    <span style='background:#E8E8E8'>pyrcc5 -o resources.py resources.qrc</span>
    
<font size = 4><i>By the command below, check if labelImg is working</i></font>

    <span style='background:#E8E8E8'>python labelImg.py</span>
    


<font size = 4><b>Note:</b> <i>If the warning says <font color = "red">ModuleNotFoundError: No module named 'libs.resources'</font>
      Copy the <b>resources.py</b> from <b>\addons</b> to <b>\addons\libs</b></i></font>
</pre>
